BREAK AND CONTINUE


```python
fruits = ["apple","banana","cherry","onion","kiwi","orange"]
for x in fruits:
    if x=="onion":
        break
    print(x)
```

```python
fruits = ["apple","banana","cherry","onion","kiwi","orange"]
for x in fruits:
    if x=="onion":
        continue
    print(x)
```
