```python
student = {
    "name": "John Doe",
    "age": 20,  
    "course": "Computer Science"
}
print(student)
```

```python
temp ={
    "lucknow": [30,35,40],
    "delhi": [25,30,35],
    "mumbai": [20,25,30]
}
print(temp)
```

```python
info = dict(name="Alice", age=25, city="New York")
print(info)
```

```python
#accessing values using keys
#1st method 
student ['name']
student ['age']
```

```python
#2nd method
student.get('Name')
student.get('age')
```

```python
print(student.keys())
print(student.values())
print(student.items())
```

```python
#adding a new pair 
student['city']= "New York"
print (student)
```

```python
#deleting a pair
del student['city']
print(student)
```

```python
#updating a pair
student['course']='BBA'
print(student)
```

```python
for i in student:
    print(i)
```

```python
for i,v in student.items():
    print(f'{i}: {v}')
```

```python
#f string ka use 
a=10
b=20
print (f"sum of {a} and {b} is {a+b}")
```

```python
#nested dictionary
company = {
    "employee1": {
        "name": "John",
        "age": 30,
        "department": "HR"
    },
    "employee2": {
        "name": "Alice",
        "age": 28,
        "department": "Finance"
    }
}
print (company)
```

```python
company['employee3'] = {
    'name': 'Raj Yadav', 
    'age': 35, 
    'department': 'IT'}
print (company)
```

```python
company['employee1']['age'] = 31
print (company)
```
