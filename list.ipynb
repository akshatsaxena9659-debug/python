INDEXING AND SLICING 


```python
apps = ["Facebook", "Instagram", "Snapchat", "Twitter", "TikTok", "YouTube", "WhatsApp"]
print(apps[1])
print(apps[5])
print(apps[-2])
print(apps[-5])
```

LIST METHODS 


- append - adds an element at the end of the list 
- insert - add an element at any specific position 
- pop - deletes an element from the end and also using index value 
- remove - deletes an element using the name of it 
- copy - make a copy of the list 
- sort - sort the elements in specific order 
- extend - combine two list 

```python
apps.append("LinkedIn")
print(apps)
```

```python
apps.insert(2, "pinterest")
print(apps)
```

```python
apps.pop() #deletes the last item in the list
print(apps)
```

```python
apps.pop(3)
print(apps)
```

```python
apps.remove("Instagram")
print(apps)
```

```python
apps2=apps.copy()
print(apps2)
print(apps)
```

```python
apps2.clear()
print(apps2)
```

```python
apps.sort() 
print(apps)
```

```python
list =[1,2,3,4,5,]
apps.extend(list)
print(apps)
```

```python
list.extend(apps)
print(list)
```
